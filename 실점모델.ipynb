### ERA(평균자책점)

```python
team['ERA']=(team['ER']*9)/(team['INN2']/3)
team['ERA']
```

### FIP (수비무관 평균자책점)

```python
team['FIP']=(team['VS_HR']*13+3*team['VS_BB']+team['VS_HP']*3-2*team['VS_KK'])/(team['INN2'])
team['FIP']
```

### KFIP (KBO리그 전용 수비무관 평균자책점)

```python
team['KFIP']=(team['VS_HR']*14+(team['VS_BB']+team['HP'])*3-team['VS_KK'])/(team['INN2'])
team['KFIP']
```

### WHIP (이닝당 출루허용률)

```python
team['WHIP']=(team['VS_HIT']+team['VS_BB']+team['VS_HP'])/(team['INN2'])
team['WHIP']
```

### LOB% (잔루처리율)

```python
team['LOB']=(team['VS_HIT']+team['VS_BB']+team['VS_HP']-team['R'])/(team['VS_HIT']+team['VS_BB']+team['VS_HP']-1.4*team['VS_HR'])
team['LOB']
```

### ERC (성분추정 평균자책점)

##### PTB

```python
PTB=((team['VS_HIT']-team['VS_HR'])*1.255+team['VS_HR']*4)+((team['VS_BB']+team['VS_HP']-team['VS_IB'])*0.56)
```

```python
team['ERC']=(((team['VS_HIT']+team['VS_BB']+team['VS_HP'])*PTB)/(team['VS_PA']*team['INN2']*3))-.056
team['ERC']
```

### kwERA (수비무관 삼진대비 평균자책점)

```python
team['kwERA']=((team['VS_KK']-(team['VS_BB']+team['VS_HP']-team['VS_IB'])*12)/(team['VS_PA']))+5.4
team['kwERA']
```

### QS (퀄리티 스타트)

```python
team['INT_INN2'] = team['INN2'] / 3
team['INT_INN2']
```

```python
team['ERA']
```

```python
team['INT_INN2'] =team['INT_INN2'].astype('int')
team['INT_ERA'] = team['ERA'].astype('int')
```

```python
conditionlist = [
    (team['INT_INN2'] >=6) & (team['ERA'] <= 3),
    (team['INT_INN2']<6),
    (team['ERA']>3)
]
choicelist = [1, 0, 0]
team['QS'] = np.select(conditionlist, choicelist)
```

### BABIP

```python
team['BABIP']=(team['VS_HIT']+team['VS_HR'])/(team['VS_AB']-team['VS_KK']-team['VS_HR']+team['VS_SF'])
team['BABIP']
```

### K/BB

```python
team['K/BB']=team['VS_BB']/team['VS_KK']
team['K/BB']
```

### 사이영 포인트

```python
team['SAP']=0.55*team['INN2']-team['ER']+(team['VS_KK']/10)
team['SAP']
```
