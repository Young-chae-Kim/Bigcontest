```python
def PS(dataset, t_id):
    
    PS = dataset[dataset['START_CK']==1]
    
    t_id = PS[PS['T_ID']== t_id]
    
    PS = t_id.groupby(['P_ID'], as_index=False).START_CK.count()
    
    return PS
```

```python
def PS_YEAR(t_id,dataset1, dataset2, dataset3, dataset4, dataset5, dataset6, dataset7, dataset8, dataset9, dataset10):
    
    PS_1 = PS(dataset1, t_id)
    PS_2 = PS(dataset2, t_id)
    PS_3 = PS(dataset3, t_id)
    PS_4 = PS(dataset4, t_id)
    PS_5 = PS(dataset5, t_id)
    PS_6 = PS(dataset6, t_id)
    PS_7 = PS(dataset7, t_id)
    PS_8 = PS(dataset8, t_id)
    PS_9 = PS(dataset9, t_id)
    PS_10 = PS(dataset10, t_id)
    
    PS_a = pd.merge(PS_1, PS_2, on=['P_ID'], how='outer')
    PS_b = pd.merge(PS_a, PS_3, on=['P_ID'], how='outer')
    PS_c = pd.merge(PS_b, PS_4, on=['P_ID'], how='outer')
    PS_d = pd.merge(PS_c, PS_5, on=['P_ID'], how='outer')
    PS_e = pd.merge(PS_d, PS_6, on=['P_ID'], how='outer')
    PS_f = pd.merge(PS_e, PS_7, on=['P_ID'], how='outer')
    PS_g = pd.merge(PS_f, PS_8, on=['P_ID'], how='outer')
    PS_h = pd.merge(PS_g, PS_9, on=['P_ID'], how='outer')
    PS_YEAR = pd.merge(PS_h, PS_10, on=['P_ID'], how='outer')
    
    pd.options.display.float_format='{:,.0f}'.format
    PS_YEAR=PS_YEAR.fillna(0)
    PS_YEAR['T_ID']=t_id
    PS_YEAR.columns = ['P_ID','12','3','23','4','34','5','45','6','56','7','T_ID']
    
    return PS_YEAR
```

```python
# 개인타자에 가중해주는 선발투수 ERA 
def PH_PS(dataset):
    
    PH = dataset[dataset['START_CK']==1]
    PH_PS = PH[['G_ID','T_ID','P_ID']]
    
    ERA_d = PH.groupby(['P_ID'], as_index=False).ER.sum()
    ERA_d['ER'] = ERA_d['ER']*9
    ERA_n = PH.groupby(['P_ID'], as_index=False).INN2.sum()
    ERA_n['INN2'] = ERA_n['INN2']/3
    ERA_NUM = pd.merge(ERA_d, ERA_n, on=['P_ID'], how='inner')
    ERA_NUM['ERA']=ERA_d['ER']/ERA_n['INN2']
    ERA_NUM=ERA_NUM.drop(['ER','INN2'],axis=1)
    
    PH_PS = pd.merge(PH_PS, ERA_NUM, on=['P_ID'], how='left')
    
    return PH_PS
```
